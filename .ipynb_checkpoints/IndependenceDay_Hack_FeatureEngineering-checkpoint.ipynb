# Exploratory Data Analysis

```python
#supressing all the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
def NumericalVariables_targetPlots(df,segment_by,target_var = "m13"):
    """A function for plotting the distribution of numerical variables and its effect on default"""
    
    fig, ax = plt.subplots(ncols= 2, figsize = (14,6))    

    #boxplot for comparison
    sns.boxplot(x = target_var, y = segment_by, data=df, ax=ax[0])
    ax[0].set_title("Comparision of " + segment_by + " vs " + target_var)
    
    #distribution plot
    ax[1].set_title("Distribution of "+segment_by)
    ax[1].set_ylabel("Frequency")
    sns.distplot(a = df[segment_by], ax=ax[1], kde=False)
    
    plt.show()
```

```python
def CategoricalVariables_targetPlots(df, segment_by,invert_axis = False, target_var = "m13"):
    
    """A function for Plotting the effect of variables(categorical data) on default """
    
    fig, ax = plt.subplots(ncols= 2, figsize = (14,6))
    
    #countplot for distribution along with target variable
    #invert axis variable helps to inter change the axis so that names of categories doesn't overlap
    if invert_axis == False:
        sns.countplot(x = segment_by, data=df,hue="m13",ax=ax[0])
    else:
        sns.countplot(y = segment_by, data=df,hue="m13",ax=ax[0])
        
    ax[0].set_title("Comparision of " + segment_by + " vs " + "m13")
    
    #plot the effect of variable on attrition
    if invert_axis == False:
        sns.barplot(x = segment_by, y = target_var ,data=df,ci=None)
        ax[1].set_ylabel("Average(default)")
    else:
        sns.barplot(y = segment_by, x = target_var ,data=df,ci=None)
        ax[1].set_xlabel("Average(default)")
        
    ax[1].set_title("default rate by {}".format(segment_by))
    
    plt.tight_layout()

    plt.show()
```

# Feature Engineering
> Create new variables

```python
traindf.head()
```

```python
#Merging both train and test data so that we can do feature engineering on both datasets at once

traindf["source"] = "train"
testdf["source"] = "test"
```

```python
#merge both 

totaldf = pd.concat([traindf, testdf], sort = False)
```

```python
totaldf.head()
```

```python
totaldf.financial_institution.value_counts()
```

```python
#amount of unpaid balance to be paid per day.

totaldf["unpaid_balance_day"] = totaldf.unpaid_principal_bal//totaldf.loan_term
```

## Extract features from 'origination_date' and 'first payment date'
- Month
- Year
- Day of the week

```python
#convert to date time format

totaldf.origination_date = pd.to_datetime(totaldf.origination_date, format='%d-%m-%Y')
```

```python
#getting the origination month, year and day of the week

totaldf["origination_month"] = totaldf.origination_date.dt.month
totaldf["origination_year"] = totaldf.origination_date.dt.year
totaldf["orignation_weekday"] = totaldf.origination_date.dt.dayofweek
```

```python
#similarly we will convert the 'first payment date' to date format and extract data from it.
totaldf.first_payment_date = pd.to_datetime(totaldf.first_payment_date, format='%d-%m-%Y')
```

```python
totaldf["first_payment_month"] = totaldf.first_payment_date.dt.month
totaldf["first_payment_year"] = totaldf.first_payment_date.dt.year
totaldf["first_payment_weekday"] = totaldf.first_payment_date.dt.dayofweek
```

```python
#calculate the time gap between the origination and first payment date

timegap = totaldf.first_payment_date - totaldf.origination_date
```

```python
totaldf["gap_days"] = timegap.dt.days
```

```python
#clean the column financial institution - removing the "," and " "

totaldf.financial_institution = totaldf.financial_institution.str.replace(",", "").str.replace(" ", "")
```

```python
#drop the date columns

totaldf.drop(["origination_date", "first_payment_date", "loan_id"], axis = 1, inplace=True)
```

```python
totaldf_onehot = totaldf.copy()
```

# Handling Categorical Variables

```python
loan_dummies = pd.get_dummies(totaldf_onehot.loan_purpose, prefix="loan")
loan_dummies.head()
```

```python
## Merging this dataframe
totaldf_onehot = pd.concat([totaldf_onehot,loan_dummies],axis=1)
totaldf_onehot.drop(['loan_purpose'], axis = 1, inplace = True)
```

```python
financial_dummies = pd.get_dummies(totaldf_onehot.financial_institution)
financial_dummies.head()
```

```python
## Merging this dataframe
totaldf_onehot = pd.concat([totaldf_onehot,financial_dummies],axis=1)
totaldf_onehot.drop(['financial_institution'], axis = 1, inplace = True)
```

```python
#final dataset

totaldf_onehot.head()
```

## Convert the Categorical Values to Pandas Category

```python
cat_var_names = [key for key in dict(totaldf.dtypes) if dict(totaldf.dtypes)[key] in ['object']]
```

```python
#convert the string data objects to category 

for col in cat_var_names:
    totaldf[col] = totaldf[col].astype('category')
```

## Memory Usage

```python
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2    
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)    
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```python
totaldf_onehot = reduce_mem_usage(totaldf_onehot)
```
